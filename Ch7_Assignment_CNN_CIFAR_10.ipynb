Part1: Import libraries

```python
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
```

Part 2: Data Preparation

```python
# Define transforms for the training and test sets
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

# Download and load the training and test datasets
trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=32,
                                          shuffle=True, num_workers=2)

testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=32,
                                         shuffle=False, num_workers=2)

# Define the classes
classes = ('plane', 'car', 'bird', 'cat', 'deer',
           'dog', 'frog', 'horse', 'ship', 'truck')
```

Visualize some training images:



```python
# Function to show an image
def imshow(img):
    img = img / 2 + 0.5  # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()

# Get some random training images
dataiter = iter(trainloader)
images, labels = next(dataiter)

# Show images
imshow(torchvision.utils.make_grid(images))
# Print labels
print(' '.join(f'{classes[labels[j]]:5s}' for j in range(4)))
```

Part 3: Building the CNN


```python
class SimpleCNN(nn.Module):  # Define a class for the CNN that inherits from nn.Module
    def __init__(self):
        super(SimpleCNN, self).__init__()  # Initialize the base class
        self.conv1 = nn.Conv2d(3, 6, 5)  # First convolutional layer (3 input channels, 6 output channels, 5x5 kernel)
        self.pool = nn.MaxPool2d(2, 2)  # Max pooling layer (2x2 window, stride 2)
        self.conv2 = nn.Conv2d(6, 16, 5)  # Second convolutional layer (6 input channels, 16 output channels, 5x5 kernel)
        self.fc1 = nn.Linear(16 * 5 * 5, 120)  # First fully connected layer (16*5*5 input features, 120 output features)
        self.fc2 = nn.Linear(120, 84)  # Second fully connected layer (120 input features, 84 output features)
        self.fc3 = nn.Linear(84, 10)  # Third fully connected layer (84 input features, 10 output features)

    def forward(self, x):  # Define the forward pass
        x = self.pool(nn.functional.relu(self.conv1(x)))  # Apply conv1, ReLU, and pooling
        x = self.pool(nn.functional.relu(self.conv2(x)))  # Apply conv2, ReLU, and pooling
        x = x.view(-1, 16 * 5 * 5)  # Flatten the tensor
        x = nn.functional.relu(self.fc1(x))  # Apply fc1 and ReLU
        x = nn.functional.relu(self.fc2(x))  # Apply fc2 and ReLU
        x = self.fc3(x)  # Apply fc3
        return x  # Return the output

net = SimpleCNN()  # Create an instance of the CNN
```

Defining the loss function and optimizer and Training the Network



```python
criterion = nn.CrossEntropyLoss()  # Define the loss function (cross-entropy loss for classification)
optimizer = torch.optim.SGD(net.parameters(), lr=0.001, momentum=0.9)  # Define the optimizer (SGD with learning rate 0.001 and momentum 0.9)

for epoch in range(2):  # Loop over the dataset multiple times (2 epochs)

    running_loss = 0.0  # Initialize the running loss
    for i, data in enumerate(trainloader, 0):  # Iterate over the training data
        inputs, labels = data  # Get the inputs and labels from the data

        optimizer.zero_grad()  # Zero the parameter gradients

        outputs = net(inputs)  # Forward pass
        loss = criterion(outputs, labels)  # Compute the loss
        loss.backward()  # Backward pass (compute gradients)
        optimizer.step()  # Optimize (update weights)

        running_loss += loss.item()  # Update the running loss
        if i % 2000 == 1999:  # Print every 2000 mini-batches
            print(f'[Epoch {epoch + 1}, Batch {i + 1}] loss: {running_loss / 2000:.3f}')  # Print the average loss
            running_loss = 0.0  # Reset the running loss

print('Finished Training')  # Indicate that training is finished
```

Part 5: Evaluating the Network


```python
correct = 0  # Initialize the count of correct predictions
total = 0  # Initialize the total number of predictions
with torch.no_grad():  # Disable gradient calculation
    for data in testloader:  # Iterate over the test data
        images, labels = data  # Get the inputs and labels
        outputs = net(images)  # Forward pass
        _, predicted = torch.max(outputs.data, 1)  # Get the predicted class
        total += labels.size(0)  # Update the total count
        correct += (predicted == labels).sum().item()  # Update the correct count

print(f'Accuracy of the network on the 10000 test images: {100 * correct // total}%')  # Print the accuracy
```

Class-wise accuracy:



```python
class_correct = list(0. for i in range(10))  # Initialize the count of correct predictions per class
class_total = list(0. for i in range(10))  # Initialize the total number of predictions per class
with torch.no_grad():  # Disable gradient calculation
    for data in testloader:  # Iterate over the test data
        images, labels = data  # Get the inputs and labels
        outputs = net(images)  # Forward pass
        _, predicted = torch.max(outputs, 1)  # Get the predicted class
        c = (predicted == labels).squeeze()  # Compare predictions to labels
        for i in range(4):  # Update counts for each class in the batch
            label = labels[i]
            class_correct[label] += c[i].item()
            class_total[label] += 1

for i in range(10):  # Print the accuracy for each class
    print(f'Accuracy of {classes[i]} : {100 * class_correct[i] // class_total[i]}%')
```
